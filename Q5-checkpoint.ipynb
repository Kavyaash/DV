```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(0)  # For reproducibility
x = np.linspace(0, 10, 100)
y = np.sin(x)
y2 = np.cos(x)
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]
data = np.random.randn(1000)

# figure size for all plots
plt.figure(figsize=(15, 12))

#Line Plot
plt.subplot(2, 3, 1) 
plt.plot(x, y, label='Sine Wave', color='blue')
plt.plot(x, y2, label='Cosine Wave', color='orange')
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid()

#Bar Plot
plt.subplot(2, 3, 2)
plt.bar(categories, values, color='green')
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')

#Scatter Plot
plt.subplot(2, 3, 3)
plt.scatter(x, y, color='red', alpha=0.5)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

#Pie Chart
plt.subplot(2, 3, 4)
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=90)#autopct to customize how the percentages are displayed.
plt.title('Pie Chart')

#Histogram
plt.subplot(2, 3, 5) 
plt.hist(data, bins=30, color='purple', alpha=0.7)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')

#Box Plot
plt.subplot(2, 3, 6)
plt.boxplot(data, vert=True, patch_artist=True)
plt.title('Box Plot')
plt.ylabel('Values')

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```
